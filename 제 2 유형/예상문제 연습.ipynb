체험환경 연습문제 풀이

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcwd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이브 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv("data/customer_train.csv")
test = pd.read_csv("data/customer_test.csv")

# print(.shape)
# print(.shape)

# 사용자 코딩
# 1. 라이브러리 및 데이터 확인
x_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
x_test = test

print(x_train.columns)
print(x_test.columns)
print("x_train : ", x_train.shape)
print("x_test : ", x_test.shape)
print("y_train : ", y_train.shape)
print(y_train.head())

# 2. 데이터 탐색
print(x_train.info())
print(x_test.info())
print(y_train.info())
print("-"*30)
# train, test 분포 정도 확인 (이상치 없는듯)
print(x_train.describe().iloc[:, 3])
print(x_test.describe().iloc[:, 3])

print("-"*30)
print(y_train.describe())
# print(train)

# 3. 데이터 전처리 및 분리
# 1) 결측치
print(round(x_train.isnull().sum() / x_train.shape[0], 2))
print("-"*30)
print(round(x_test.isnull().sum() / x_test.shape[0], 2))
print("-"*30)
print(round(y_train.isnull().sum() / y_train.shape[0], 2))

# train, test 데이터셋의 환불금액 컬럼 결측치가 60%이므로 중앙값으로 대체 처리 => 중앙값이 아니라 0으로 대체해야함!!!
# print("train 환불금액 중앙값 : ", x_train['환불금액'].median())
x_train['환불금액'] = x_train['환불금액'].fillna(0)
x_test['환불금액'] = x_test['환불금액'].fillna(0)

print("-"*30)
print(round(train.isnull().sum() / train.shape[0], 2))
print(round(test.isnull().sum() / test.shape[0], 2))

# # 2) 이상치 : describe() 함수로 컬럼별로 확인했을 때 이상치 없어보임.

# 3) 변수 처리 : 총 구매건수 컬럼 추가?
# 3-1) 학습에 필요 없는 회원 ID 삭제
x_train = x_train.drop('회원ID', axis=1)
x_test = x_test.drop('회원ID', axis=1)

# 3-2) 원핫인코딩
print("-"*30)
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)
print(x_train.shape)
print("-"*30)
print(x_test.shape)
print("-"*30)
# 원핫인코딩 후 컬럼이 안맞으므로 맞춰주자.
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)

# # 4) 데이터 분리
print("-"*30)
# print(train.iloc[:, :-1].shape)
# print(train.iloc[:, -1].shape)
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, stratify=y_train, random_state=2024)
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)

# 4. 모델링 및 평가
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)				# 학습
y_pred = model.predict(x_val)			# 검증용 예측
print(y_pred)

# # roc-auc 평가 지표로 평가
from sklearn.metrics import roc_auc_score
roc = roc_auc_score(y_val, y_pred)
print(roc)


# 5. 예측값 제출
y_result = model.predict(x_test)
print(y_result[:5])
result = pd.DataFrame({'pred' : y_result})
result.to_csv("result.csv", index=False)

df = pd.read_csv("result.csv")
print(df.head())



# 답안 제출 참고
# 아래 코드는 예시이며 변수명 등 개인별로 변경하여 활용
# pd.DataFrame변수.to_csv("result.csv", index=False)
```
